```python
def extract_article_text(url):
    """Extracts article content using newspaper3k given a URL."""
    article = Article(url)
    try:
        article.download()
        article.parse()
        # Optionally you can call article.nlp() here if you need summary, keywords, etc.
        return article.text
    except Exception as e:
        print(f"Error processing URL: {url}\nError: {e}")
        return None
```

```python
# Function to clean the text
def clean_text(text):
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove special characters and numbers (optional, depending on your needs)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    return text

# Apply the cleaning function to the 'text_content' column
df['cleaned_text'] = df['text_content'].astype(str).apply(clean_text)
```

```python
def truncate_text(text, max_length=510):
    """Truncates text to the specified maximum length."""
    # Tokenize the text and truncate to max_length
    tokens = tokenizer.tokenize(text)
    truncated_tokens = tokens[:max_length]
    # Convert tokens back to text
    truncated_text = tokenizer.convert_tokens_to_string(truncated_tokens)
    return truncated_text
```
