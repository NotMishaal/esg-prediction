```python
import pandas as pd
import numpy as np
import re
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline

# Environmental model
name = "ESGBERT/EnvironmentalBERT-environmental"
tokenizer = AutoTokenizer.from_pretrained(name)
model = AutoModelForSequenceClassification.from_pretrained(name)
pipe_env = pipeline("text-classification", model=model, tokenizer=tokenizer)

# Social model
name = "ESGBERT/SocialBERT-social"
tokenizer = AutoTokenizer.from_pretrained(name)
model = AutoModelForSequenceClassification.from_pretrained(name)
pipe_soc = pipeline("text-classification", model=model, tokenizer=tokenizer)

# Governance model
name = "ESGBERT/GovernanceBERT-governance"
tokenizer = AutoTokenizer.from_pretrained(name)
model = AutoModelForSequenceClassification.from_pretrained(name)
pipe_gov = pipeline("text-classification", model=model, tokenizer=tokenizer)
```

```python
data = pd.read_csv("clean_news.csv")
```

```python
for index, row in df.iterrows():
  env = pipe_env(df['cleaned_text'][index], padding=True, truncation=True)
  soc = pipe_soc(df['cleaned_text'][index], padding=True, truncation=True)
  gov = pipe_gov(df['cleaned_text'][index], padding=True, truncation=True)

  env_labels = [x["label"] for x in env]
  soc_labels = [x["label"] for x in soc]
  gov_labels = [x["label"] for x in gov]

  data_env = pd.DataFrame({"sentence": df['cleaned_text'][index], "environmental": env_labels, "social": soc_labels, "governance": gov_labels})
  # break
  for in_index, in_row in data_env.iterrows():
    # Check if any of the ESG columns have a value other than "none"
    if in_row['environmental'] != 'none' or in_row['social'] != 'none' or in_row['governance'] != 'none':
        print(f"Row {in_index}: ESG relevance found.")
    else:
        print(f"Row {in_index}: No ESG relevance found.")
        df['cleaned_text'][index].pop(in_index)
```

```python
# Calculate ratios
total_count = len(data_env)
environmental_ratio = data_env[data_env["environmental"] == "environmental"].shape[0] / total_count
social_ratio = data_env[data_env["social"] == "social"].shape[0] / total_count
governance_ratio = data_env[data_env["governance"] == "governance"].shape[0] / total_count
none_environmental_ratio = 1 - environmental_ratio
none_social_ratio = 1- social_ratio
none_governance_ratio = 1 - governance_ratio

print(f"Environmental Ratio: {environmental_ratio}")
print(f"Social Ratio: {social_ratio}")
print(f"Governance Ratio: {governance_ratio}")
print(f"None Environmental Ratio: {none_environmental_ratio}")
print(f"None Social Ratio: {none_social_ratio}")
print(f"None Governance Ratio: {none_governance_ratio}")
```
